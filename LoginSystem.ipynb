```python
import xlrd
import xlwt 
from xlwt import Workbook 

class login:

    def __init__(self):
        print("!!!!!!!!!!!!!Welcome to the Registration & Login Screen!!!!!!!!!!!!!")
        
        # Loading existing data in Excel to Dictionary
        self.d={}
        self.loginDetails=xlrd.open_workbook('LoginDetails.xls')
        self.sheet=self.loginDetails.sheet_by_index(0)
        self.numberOfIDs=self.sheet.nrows
        for i in range(self.numberOfIDs):
            self.ExistingloginID=self.sheet.cell_value(i,0)
            self.ExistingloginPWD=self.sheet.cell_value(i,1)
            self.d[self.ExistingloginID] = self.ExistingloginPWD
        
    def userInput(self):
        # Accepting Main Menu Option. Validating for Correct Option.
        self.validInput=None
        print('\nEnter 1 for Registration')
        print('Enter 2 for User Login')
        print('Enter 3 for Forgot Password')
        self.option=input()

        if self.option.isdigit()==False or int(self.option)> 3:
            print('Invalid Option. Pls retry again')
            self.validInput=False
        else:
            self.validInput=True
            self.option=int(self.option)
            
        return self.validInput,self.option
            
        
    def userLogin(self):
        # Accepting user login details
        self.loginID=input("\nEnter the UserID   :  ")
        self.loginPWD=input("Enter your Password:  ")  
        
    def existingUserValidation(self):     
        # Validating the User entered USER ID and PWD against the Dictionary items
        self.idValidate=None
        for self.keys,self.values in self.d.items():
            if self.keys==self.loginID:
                self.idValidate='ExistingUser'
                if self.values==self.loginPWD:
                    self.idValidate = 'allGood'
                    print('\nYou logged on SUCCESSFULLY.. Welcome to the Python World!!')
                else:
                    self.idValidate='IncorrectPWD'
                    print('\nIncorrect Password, Please retry again!!!')
    
        return self.idValidate
        
    def register(self):
        #Getting new User ID and Password for Registration
        self.newLoginID=input('\nEnter the USER ID for Registration :  ')
        self.newLoginPWD=input('Enter the Password                 :  ') 

    def newUserIDValidation(self):
        #Validating new UserID with all requirements
        import re
        self.count1=0
        self.count2=0
        self.position=None
        self.validID=None
        
        self.l1=re.findall('@',self.newLoginID)
        self.count1 = len(self.l1)
        
        if self.count1==1:
            for match in re.finditer('@',self.newLoginID):
                self.position=match.start()
                
            self.l2=re.split('@',self.newLoginID,2)
            
            self.count2=str(self.l2[1]).count('.')
            
        if self.newLoginID[0].isalpha() == False or self.count1 != 1 or self.count2 <1 or self.newLoginID[self.position+1]=='.':
            self.validID=False
            print('\nInvalid USER ID')
        else:
            self.validID=True
            #validating if its already existing in database. 
            for self.keys,self.values in self.d.items():
                if self.keys==self.newLoginID:
                    self.validID=False
                    print('\nUserID already in Database. Pls Register with other UserID')
                           
        return self.validID
    
    def newUserPWDvalidation(self):
        self.validPWD=None
        self.specialChar='[@_!#$%^&*()<>?/\|}{~:]'
        self.upperCaseCount=0
        self.lowerCaseCount=0
        self.digitCount=0
        self.specialCharCount=0
        
        #Validating new UserID with all requirements
        self.len1=len(self.newLoginPWD)
        
        for c in self.newLoginPWD:
            if c.isupper():
                self.upperCaseCount+=1
            elif c.islower():
                self.lowerCaseCount+=1
            elif c.isdigit():
                self.digitCount+=1
            elif c in self.specialChar:
                self.specialCharCount+=1
            else:
                pass
        
        if (5>len(self.newLoginPWD)>16) or (self.upperCaseCount < 1) or (self.lowerCaseCount < 1) or (self.digitCount < 1)  or (self.specialCharCount < 1):
            print('\nPassword must follow below requirements.')
            print('Length MUST be between 6 to 15.')
            print('MUST have one special Character, one Digit, one Uppercase and one Lowercase letter.')
            self.validPWD=False
        else:
            self.validPWD=True
            
        return self.validPWD
    
    def updateInRegister(self):
        #Validating the New UserID and Password before inserting into Database
        # Workbook is created 
        wb = Workbook() 
        sheet1 = wb.add_sheet('Sheet 1') 
        
        self.d[self.newLoginID] = self.newLoginPWD

        row = 0
        #iterate the each key-value pair of dictionary & insert into sheet
        for k, v in self.d.items():
            sheet1.write(row,0,k) 
            sheet1.write(row,1,v) 
            row = row + 1

        wb.save('LoginDetails.xls')
        print(f"\nRegistration is SUCCESSFULL for {self.newLoginID}")
    
    def fetchPassword(self):
        #Fetching the Password for the UserID
        self.fetchingUserID=input("\nEnter your existing UserID  : ")
        self.fetchUserCheck=False

        for self.keys,self.values in self.d.items():
            if self.keys==self.fetchingUserID:
                self.fetchUserCheck = True
                print(f"\nThe password for UserID {self.fetchingUserID} is {self.values}")
                
        if self.fetchUserCheck == False:
            print("\nUserID does not Exist in Database. Pls try with Existing UserID!!!!!")
            
        return self.fetchUserCheck
```
